### Exercise 08.1


The aim of the first exercise is to write a code able to determine the expectetion value of 

$$
\langle {\hat H} \rangle_T = 
\frac{\int dx \Psi^*_T(x) {\hat H} \Psi_T(x)}
{\int dx |\Psi_T(x)|^2} = 
\int dx \frac{|\Psi_T(x)|^2}{\int dx |\Psi_T(x)|^2} \frac{{\hat H} \Psi_T(x)}{\Psi_T(x)}
$$

To do so one need to first calculate the normalisation of the state, which is $$\int dx |\Psi_T(x)|^2 = 2 \sqrt {\pi}\sigma(1+e^{\mu^2/\sigma^2}) $$ 

Then the quantity $\hat H |\Psi_T(x)\rangle$ is evaluated as 
$$\Psi_T(x) [\frac{-\hbar^2}{2m\sigma^4}(x^2 + \mu^2 - \sigma^2 - 2x\mu\tanh \frac{\mu x}{\sigma^2}) + V(x)]$$

This is done through the Metropolis algorithm: a new position is randomly chosen from an interval around the previous position, which is then accepted with a probability given by the wavefunction. The energy of the new position (if the step was accepted) or of the old one (if it isn't) is then computed, and the process starts again.

Then $|\Psi_T(x)|^2|$ is sampled using the Metropolis algorithm.
The results are mediated over $10^6$ steps to find the average value, and the uncertainties are computed using the blocking method. 

The system reaches the equilibrium quickly, as seen in the second plot, where the X coordinate is plotted. Starting from X = 20 it arrives in a neighbourhood of 0 within approximately 20 steps. The code was tested with $\sigma = 0.8$ and $\mu = 0.6$, and the steps are taken within an interval of 3.3 in either direction to have an acceptance rate of ~50%.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

fig=plt.figure(figsize=(15,10))

x,y, y_err=np.loadtxt("ExpectationValue.dat", usecols=(0, 1, 2), skiprows = 0, unpack=True)
plt.title("Expectation Value", fontweight="bold")
plt.xlabel("Number of blocks")
plt.ylabel("Average value")
plt.errorbar(x, y, yerr=y_err, color = "blue")
plt.grid(True)
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

def Vpot(x):
    return (x**2 - 2.5)*x**2
    #return 0.5*x**2

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)

print("Ground state energy: ", E[0])

# Plot a few things
plt.figure(figsize=(16,9))
scale = 0.3
plt.plot(x,(psi[0])**2, label = "Analitycal $|\Psi_T(x)|^2$", color = "midnightblue")

xMC=np.loadtxt("Histogram.dat", unpack=True)
plt.hist(xMC,100, (-3,3), density=True, color="darkcyan", label = "Sampled $|\Psi_T(x)|^2$")
plt.title("Sampled and Analitycal Distributions")
plt.xlabel("x")
plt.grid(True)
plt.xlim((-3,3))
plt.ylim((-0.2,0.6))
plt.legend()
plt.show()
```

### Exercise 8.03

In this exercise PIGS (Path Integral Ground State) method is used to approximate the Ground State Waverfunction  utilising the imaginary times propagation. Two different wavefunctions have been used as input:
- $\Psi_T(x)=1$
- $\Psi_T(x) = \Psi_T^{\sigma,\mu}(x)$ with the optimized variational parameters.

To obtain a similar compatibility using the two different input wavefunctions different imaginary times had to be used, specifically $\tau$ = 12 fot the constant wavefunction and $\tau$ = 8, 4 and 1 for the optimized one. The comparisons between the analytical Ground State wavefunction and the other two are shown below.

Then the differences between the Ground State estimations found using the optimized wavefunction with $\tau$ = 1, 4 and 8 are graphed, showing how changing $\tau$ affects the precision on the estimation of the actual Ground State.

```python
N = 400 # number of iterations
# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)
CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

Em,psi = np.linalg.eigh(H)
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)


fig=plt.figure(figsize=(20,20))
fig.suptitle('PIGS', fontsize=20, fontweight="bold")

plt.subplot(221)
x,k,kerr=np.loadtxt("QMC_1D/test/QMC_1D/kinetic.dat", usecols=(0, 1, 2), unpack=True)
v,verr=np.loadtxt("QMC_1D/test/QMC_1D/potential.dat", usecols=(1, 2), unpack=True)
E=k+v
Eerr =np.sqrt(kerr*kerr+verr*verr)
x= range(E.size)
plt.plot(x, E, color="goldenrod", label = "Energy PIGS", linewidth=0.5)
plt.hlines(Em[0], x[0], x[-1], color = "b", label = 'Energy Matrix Method')
plt.hlines(Energy[-1], x[0], x[-1], color = 'black', label='Energy VMC')
plt.errorbar(x, v, verr, label = "Potential PIGS", color = 'r')
plt.title("Energy PIGS - Constant Function", fontweight="bold")
plt.xlabel("M")
plt.ylabel("Energy")
plt.legend(loc = 'best')
plt.grid(True)

plt.subplot(222)
x,prob,perr=np.loadtxt("QMC_1D/test/QMC_1D/probability.dat", usecols=(0, 1, 2), unpack=True) 
plt.errorbar(x, prob - psi[0]**2, perr, label = "Constant WF T=12", color = 'goldenrod')
plt.hlines(0, x[0], x[-1], label = "Ground State", color = "black")
plt.title("GS Functions comparison - Constant", fontweight="bold")
plt.xlabel("x")
plt.ylabel("$|\psi|^2$")
plt.legend()

plt.grid(True)

plt.subplot(223)
x,k,kerr=np.loadtxt("QMC_1D/ResultsVariationalWF/QMC_1D/kinetic.dat", usecols=(0, 1, 2), unpack=True)
v,verr=np.loadtxt("QMC_1D/ResultsVariationalWF/QMC_1D/potential.dat", usecols=(1, 2), unpack=True)
E=k+v
Eerr =np.sqrt(kerr*kerr+verr*verr)
x= range(E.size)
plt.plot(x, E, color="goldenrod", label = "Energy PIGS", linewidth=0.5)
plt.hlines(Em[0], x[0], x[-1], color = "b", label = 'Energy Matrix Method')
plt.hlines(Energy[-1], x[0], x[-1], color = 'black', label='Energy VMC')
plt.errorbar(x, v, verr, label = "Potential PIGS", color ='r')
plt.title("Energy PIGS - VMC function", fontweight="bold")
plt.xlabel("M")
plt.ylabel("Energy")
plt.legend()
plt.grid(True)

plt.subplot(224)
z = np.linspace(-a/2, a/2, N)
x,prob8,perr8=np.loadtxt("QMC_1D/ResultsVariationalWF/QMC_1D/probability.dat", usecols=(0, 1, 2), unpack=True)
plt.errorbar(x, prob8-((psi[0])**2), perr8, label="VariationalMC WF T=8", color = 'goldenrod')
plt.hlines(0, x[0], x[-1], label = "Ground State", color = 'black')
plt.title("GS Functions comparison - VMC", fontweight="bold")
plt.xlabel("x")
plt.ylabel("$|\psi|^2$")
plt.grid(True)
plt.legend()

plt.show()
```
