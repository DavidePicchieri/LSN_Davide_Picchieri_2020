```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math


y_kin=np.loadtxt("output_ekin.dat", unpack=True)
y_pot=np.loadtxt("output_epot.dat", unpack=True)
y_tot=np.loadtxt("output_etot.dat", unpack=True)

x = np.arange(y_tot.size)

fig=plt.figure(figsize=(10, 7))

plt.plot(x, y_kin, label="Kinetic Energy", color = "lightblue")
plt.plot(x, y_pot, label="Potential Energy", color = "red")
plt.plot(x, y_tot, label="Total Energy", color = "purple")
plt.xlabel('Number of blocks')
plt.ylabel('$E$')
plt.title("Energy over time")
plt.grid(True)
plt.legend(loc="upper right", prop={'size': 15})

plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math


y=np.loadtxt("output_temp.dat", unpack=True)
x = np.arange(y.size)

plt.plot(x, y, label="Temperature")
plt.hlines(1.9, 0, 10000)
plt.xlabel('Number of blocks')
plt.ylabel('Temperature')
plt.title("Temperature")
plt.grid(True)
plt.legend(loc="upper right", prop={'size': 15})

plt.show()
```
