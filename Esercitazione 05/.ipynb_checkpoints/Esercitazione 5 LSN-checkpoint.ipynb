### Exercise 05.1

This exercise consists in sampling the orbital of an electron for two different energy levels of the hydrogen atom using the Metropolis algorithm. The probability distribution functions used were the Uniform distribution and the Gaussian distribution for both $|\Psi_{1,0,0}(x,y,z)|^2$ and $|\Psi_{2,1,0}(x,y,z)|^2$. 



#### Equilibration 
The initial point of the sampling sequence is chosen far away from the atom's nucleus, 50 Bohr radii away in each coordinate, and as it can be seen in the first two graphs showing the equilibration phase the points sampled get within a few radii from the nucleus after about 400 steps, but the equilibration was done with 1000 steps to be safe.

The step length was set differently for every wavefunction and probability distribution, specifically:
-$|\Psi_{1,0,0}(x,y,z)|^2$ with Uniform Distribution sampling: 1.2
-$|\Psi_{1,0,0}(x,y,z)|^2$ with Gaussian Distribution sampling: 0.75
-$|\Psi_{2,1,0}(x,y,z)|^2$ with Uniform Distribution sampling: 3
-$|\Psi_{2,1,0}(x,y,z)|^2$ with Gaussian Distribution sampling: 1.8

These values were chosen to obtaine an acceptance rate of ~50%

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D


X,Y,Z=np.loadtxt("Coords100.dat", max_rows=1000, usecols=(0, 1, 2), unpack=True)

fig = plt.figure(figsize=(16,8))
ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=X, marker='.')
ax.set_xlabel('x [$a_0$]')
ax.set_ylabel('y [$a_0$]')
ax.set_zlabel('z [$a_0$]')
ax.set_title('Equilibration for $|\Psi_{1,0,0}(x,y,z)|^2$', fontsize=18)
ax.view_init(10, 20)
plt.show()
```

#### Simulation and Results

The Simulation was run for $10^6$ steps for each wavefunction and each probability distribution.
The medium distance from the nucleus is plotted below and are consistent with the known values for both $|\Psi_{1,0,0}(x,y,z)|^2$ and $|\Psi_{2,1,0}(x,y,z)|^2$, shown with the straight horizontal line, using either distribution function for the sampling. 

The uncertainties were estimated as usual using the blocking method, since the points are correlated. The block size utilized was of 100 steps.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x_100_u,y_100_u,err_100_u=np.loadtxt("R_100_Unif.dat", usecols=(0, 1, 2), unpack=True)
x_100_g,y_100_g,err_100_g=np.loadtxt("R_100_Gauss.dat", usecols=(0, 1, 2), unpack=True)


fig=plt.figure(figsize=(16, 8))

plt.hlines(1.5, 0, 100)
plt.errorbar(x_100_u,y_100_u,yerr=err_100_u, label="Uniform Sampling")
plt.errorbar(x_100_g,y_100_g,yerr=err_100_g, label="Gaussian Sampling")
plt.xlabel('Number of Blocks', fontsize=20)
plt.ylabel('<r> [$a_0$]', fontsize=20)
plt.title("$\psi_{100}$" , fontsize=38)

plt.legend(loc="upper left", prop={'size': 15})

plt.grid(True)

plt.show()
```
