### Exercise 09.1

The aim of the exercise is to solve the Traveling Salesman Problem utilising a Genetic Algorithm. The TSP consists in finding the shortest trip connecting all of the cities. Here the problem is solved with a Genetic Algorithm, which is a stochastic algorithm. It uses 'mutations' and 'crossovers' among individuals, which here are the trips, to explore many different possibilities. Then a 'selection' keeps the best individuals in the populations to then start again. The mutations consists in swapping two cities or two group of cities, inverting the order of a sequence of cities and shifting a group of cities. The crossover happens between two trips, which exchange the order in which they visit the cities for a portion of their trip. The selection is done by elevating a random number $[0,1)$ to a power so that the resulting number will be skewed towards zero. Then the number is multiplied by the size of the population and so it gives which of the trips, ordered in ascending order by their length, will be chosen. Two different configurations have been tested: 32 cities randomly placed on a circle or inside a square. $L^{(1)}$ was chosen as the defining legth of a trip parameter. 

#### Circle

The graph of the best path found for the circle clearly shows that it is the actual best trip, since any deviation from this path would result in a crossing and therefore lengthening it. The length of the best trip is 6.24007 and it was achieved at the 524th generation. The graph at the bottom shows the progression of $L^{(1)}_{BEST}$ and $\langle L^{(1)}_{50 best} \rangle$ as the generations go on. The graph is stopped after 800 generations to make it more legible.


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

fig=plt.figure(figsize=(10,10))

x,y=np.loadtxt("CircleBestTrip.out", usecols=(1, 2), unpack=True)
plt.plot(x, y, '-o', color="Red", linewidth=1)
plt.title('Best Path',  fontweight="bold")
plt.xlabel("X")
plt.ylabel("Y")
circle = plt.Circle((0, 0), 1, color='b', fill=False)
plt.gca().add_patch(circle)

fig=plt.figure(figsize=(15,10))

x,y=np.loadtxt("AveTripGenerationCircle.out", usecols=(0, 1), max_rows = 800, unpack=True)
plt.title("Average and Best Trip Length", fontweight="bold")
plt.plot(x,y, color="r", label="Average")
yB=np.loadtxt("BestTripGenerationCircle.out",  max_rows = 800, unpack=True)
xB=np.arange(yB.size)
plt.xlabel("Number of Generations")
plt.ylabel("Length")
plt.plot(xB,yB, color="midnightblue", label="Best")
plt.grid(True)
plt.legend()
plt.show()
```
