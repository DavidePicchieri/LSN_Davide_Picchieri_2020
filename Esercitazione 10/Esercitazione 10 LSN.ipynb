### Exercise 10.1


In this exercise the Traveling Salesmen Problem is solved using the Simulated Annealing technique. The SA algorithm is a metaeuristic algorithm which takes inspiration from the actual annealing of metals. It minimizes a selected quantity, which in this case is the length of the path, utilising the Boltzmann probability distribution  with that quantity in place of the energy ($P = \frac{e^{-\beta L}} {Z}$).   

The starting temperature utilised is 10, the final temperature is 0.001 and the cooling happens at a rate of $T_{i+1} = T_{i} \times 0.99999$. By cooling the temperature this slowly there is no need to perform an equilibration every time the temperature is changed, as the system is already at equilibrium. The placement of the 32 cities is the same as in Exercise 9.1 for both the cirle and square configuration.

#### Circle

The best path was found by the SA algorithm is the same as the Genetic algorithm, with a length of 6.24007, and it is the actual best one: while the GA took 524 generations with a population of 100 individuals to find it, the SA algorithm took about $7\times10^5$ steps with only one individual.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math


fig=plt.figure(figsize=(20,8))
fig.suptitle('CIRCLE', fontweight = 'bold', fontsize = 20)
plt.subplot(121)
y=np.loadtxt("10.1/BestTripGenerationCircle_SA.out", unpack=True)
x=np.arange(y.size)
plt.title("Best Trip Length Simulated Annealing", fontweight="bold")
plt.xlabel("Number of Steps")
plt.ylabel("Length")
plt.plot(x,y, color="darkcyan", label="Simulated Annealing")
plt.grid(True)
plt.legend()

plt.subplot(122)
y=np.loadtxt("../Esercitazione 09/9.1/BestTripGenerationCircle.out", unpack=True)
x=np.arange(y.size)
plt.title("Trip Length Genetic Algorithm", fontweight="bold")
plt.xlabel("Number of Generations")
plt.ylabel("Length")
plt.plot(x,y, color="darkcyan", label="Genetic Algorithm")
plt.grid(True)
plt.legend()
plt.show()

fig=plt.figure(figsize=(20,8))
plt.subplot(121)

x,y=np.loadtxt("10.1/CircleBestTrip_SA.out", usecols=(1, 2), unpack=True)
plt.plot(x, y, '-o', color="red", linewidth=1)
plt.title('Best Path Simulated Annealing',  fontweight="bold")
plt.xlabel("X")
plt.ylabel("Y")
circle = plt.Circle((0, 0), 1, color='b', fill=False)
plt.gca().add_patch(circle)


plt.subplot(122)
x,y=np.loadtxt("../Esercitazione 09/9.1/CircleBestTrip.out", usecols=(1, 2), unpack=True)
plt.plot(x, y, '-o', color="red", linewidth=1)
plt.title('Best Path Genetic Algorithm',  fontweight="bold")
plt.xlabel("X")
plt.ylabel("Y")
circle = plt.Circle((0, 0), 1, color='b', fill=False)
plt.gca().add_patch(circle)

plt.show()
```

### Exercise 10.2

The second exercise consists in using the MPI libraries to compute the Genetic Algorithm on 4 different *Continents* and exchanging the best trip every so often. In each *Continent* the selection proceeds indipendently until a migration occurs, so that the best traits found on each *Continent* can merge. This allows for a quicker evolution process and it should help prevent for the algorithm to reach a local minimum and get stuck there, as the four cores explore more of the possibilitie.

The best result obtained with this method is pictured below, and corresponds to a length of 9.39889, which is a marked improvement from the GA alone but isn't quite as good as the one found using the Simulated Annealing algorithm. Below are the paths obtained using the GA algorithm with and without parallel computing side by side. At the bottom are shown the best trip for each continent at the end of the simulation. Two of them are the same, which correspond to the best path: the other two are longer.

```python
fig=plt.figure(figsize=(20,8))
fig.suptitle('BEST PATH IN THE SQUARE', fontweight = 'bold', fontsize = 20)
plt.subplot(121)

x,y=np.loadtxt("..//Esercitazione 09/9.1/SquareBestTrip.out", usecols=(1, 2), unpack=True)
plt.plot(x, y, '-o', color="blue", linewidth=1)
plt.title('Genetic Algorithm',  fontweight="bold")
plt.xlabel("X")
plt.ylabel("Y")

plt.subplot(122)
x,y=np.loadtxt("10.2/Square_BestPath.out", usecols=(1, 2), unpack=True)
plt.plot(x, y, '-o', color="blue", linewidth=1)
plt.title('Genetic Algorithm with Parallel Computing',  fontweight="bold")
plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```
