### Exercise 03.1

This exercise consists in estimating the European Put-Option and Call-Option prices. According to Black-Scholes one can do so assuming that the prices exhibit geometric Brownian motion (GBM). In this model the drift $\mu$ is the constant risk-free interest rate $r$, and the volatility $\sigma$ incapsulates the stochastic contribution to the time evolution of the price. 
The prices are estimated through a direct and a discrete approach.

#### Direct Sampling

The simpler of the two methods, it is enough to evaluate $$S(t) = S(0) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)t+\sigma W(t)\right]$$ at the delivery time T, with $W(t)$ ~ $\mathcal{N}(0, t)$ 

#### Discrete Sampling

This method works by subdiving the time to the delivery time in $N$ intervals, with $N$ = 100 in this exercise, and evaluating the price at the time $t_{i+1}$ using the price at the preceeding time step, using the formula $$S(t_{i+1}) = S(t_i) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)+\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$$ with $Z_i$ ~ $\mathcal{N}(0,1)$

#### Cost of the Options

The cost of the option is then simply obtained from the estimation of the price at the time of delivery, with the Put-Option being $\max [0,K - S(T)] \exp(-rT)$ and the Call-Option being $\max [0,S(T) - K] \exp(-rT)$. The factor $\exp(-rT)$ is there to account fot the gain one would have from simply keeping the money in a bank with fixed interest rate $r$.

#### Results 

A Monte Carlo simulation was run $10^6$ using the following parameters: 
- asset price at $t=0$: $S(0)=100$
- delivery time: $T=1$
- strike price: $K=100$
- risk-free interest rate: $r=0.1$
- volatility: $\sigma=0.25$

The results for the Call-Option and Put-Option estimations are shown in the 4 graphs below, with the horizantal line showing the analytic solution to the problem for comparison. The Monte Carlo simulated prices are all in good agreement with the analitycal solution, mostly within one $\sigma$ of each other. The uncertainties are obtained by using a blocking average with $10^2$ blocks of $10^4$ evaluations.


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

#Call option w/ irect estimation of S(T)
x_direct,y_direct,err_direct=np.loadtxt("Call_dir.dat", usecols=(0, 1, 2), unpack=True)

fig=plt.figure(figsize=(16, 12))
fig.suptitle('CALL OPTION AND PUT OPTION COST', fontweight = 'bold', fontsize = 20)
plt.subplot(221)
plt.hlines(14.975790778311286, 0, 100, color = 'black')
plt.errorbar(x_direct,y_direct,yerr=err_direct, color = 'b')
plt.xlabel('Block number')
plt.ylabel('Call option cost')
plt.title("Call option cost estimation using direct S(T)")
plt.grid(True)

#Call option w/ discrete estimation of the various S(t_i)
x_discrete,y_discrete,err_discrete=np.loadtxt("Call_disc.dat", usecols=(0, 1, 2), unpack=True)

plt.subplot(222)
plt.hlines(14.975790778311286, 0, 100, color = 'black')
plt.errorbar(x_discrete,y_discrete,yerr=err_discrete, color = 'b')
plt.xlabel('Block number')
plt.ylabel('Call option cost')
plt.title("Call option cost estimation using discrete S(t$_i$)")
plt.grid(True)

#Put option w/ direct estimation of S(T)
x_P_direct,y_P_direct,err_P_direct=np.loadtxt("Put_dir.dat", usecols=(0, 1, 2), unpack=True)

plt.subplot(223)
plt.hlines(5.4595325819072364, 0, 100, color = 'black')
plt.errorbar(x_P_direct,y_P_direct,yerr=err_P_direct, color = 'b')
plt.xlabel('Block number')
plt.ylabel('Put option cost')
plt.title("Put option cost estimation using direct S(T)")
plt.grid(True)

#Put option w/ discrete estimation of the various S(t_i)
x_P_discrete,y_P_discrete,err_P_discrete=np.loadtxt("Put_disc.dat", usecols=(0, 1, 2), unpack=True)

plt.subplot(224)
plt.hlines(5.4595325819072364, 0, 100, color = 'black')
plt.errorbar(x_P_discrete,y_P_discrete,yerr=err_P_discrete, color = 'b')
plt.xlabel('Block number')
plt.ylabel('Put option cost')
plt.title("Put option cost estimation using discrete S(t$_i$)")
plt.grid(True)


plt.show()
```
